```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python

ps2_data = pd.read_stata("http://fmwww.bc.edu/ec-p/data/Hayashi/nerlove63.dta")

print(ps2_data.head(3))
```

```python
ps2_data=np.log(ps2_data)
ps2_data['const'] = 1
```

```python
ps2_data.head(3)
```

```python
X = ps2_data.iloc[:, 1:]
Y=ps2_data.iloc[:, 0]
```

```python
estim1=np.linalg.inv(X.T@X)@X.T@Y
out1=np.round(estim1,2)
out1
```
